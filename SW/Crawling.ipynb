```python
from selenium import webdriver
from urllib.parse import quote_plus
from urllib.request import urlopen
import os
from selenium.webdriver.common.by import By
import requests
import time

MAX_IMAGES = 20

def make_url(search_term):
    # 네이버 이미지 검색
    base_url = 'https://search.naver.com/search.naver?sm=tab_hty.top&where=image&query='
    return base_url + quote_plus(search_term)

def create_folder_if_not_exists(directory):
    try:
        if not os.path.exists(directory):
            os.makedirs(directory)
    except OSError:
        print('Error: Creating directory. ' + directory) 
        
        
# def save_images(images, save_path):
#     for index, image in enumerate(images[:500]):
#         time.sleep(1)
#         src = image.get_attribute('src')
#         if src.startswith('data:image'):
#             continue
            
#         img_data = requests.get(src).content
#         with open(os.path.join(save_path, str(index + 1) + ".jpg"), 'wb') as handler:
#             handler.write(img_data)
#             print("img save " + save_path + str(index + 1) + ".jpg")

def save_images(images, save_path):
    img_urls = []  # 이미지 URL 리스트 생성
    for index, image in enumerate(images[:500]):  # images[:크롤링하고 싶은 사진 개수]
        src = image.get_attribute('src')
        if src not in img_urls:  # 이미지 URL이 리스트에 없을 경우에만 다운로드
            t = urlopen(src).read()
            file = open(os.path.join(save_path, str(index + 1) + ".jpg"), "wb")
            file.write(t)
            print("img save " + save_path + str(index + 1) + ".jpg")
            img_urls.append(src)  # 이미지 URL을 리스트에 추가
        
def crawl_images(search_term):
    # URL 생성
    url = make_url(search_term)
 
    # chrome 브라우저 열기
    browser = webdriver.Chrome('chromedriver')
    browser.implicitly_wait(3)  # 브라우저를 오픈할 때 시간간격을 준다.
    browser.get(url)
 
    # 이미지 긁어오기
    images = browser.find_elements(By.CSS_SELECTOR, "._image._listImage")
    num_images = len(images)
    
    save_path = "C:\\Users\\USER\\OneDrive - 목원대학교\\바탕 화면\\Play Data\\images\\" + search_term + "\\"
    create_folder_if_not_exists(save_path)
    
    images = []
    for _ in range(20):
        images.extend(browser.find_elements(By.CLASS_NAME, "_image._listImage"))
        browser.execute_script("window.scrollBy(0, 1000);")
        time.sleep(3)
 
    # 저장 경로 설정
    save_path = "C:\\Users\\USER\\OneDrive - 목원대학교\\바탕 화면\\Play Data\\images\\" + search_term + "\\"
    create_folder_if_not_exists(save_path)
 
    # 이미지 저장
    save_images(images, save_path)
 
    # 마무리
    print(search_term + " 저장 성공")
    browser.close()
    # 저장 경로 설정
 
 
    # 마무리
    print(search_term + " 저장 성공")
    browser.close()
 
if __name__ == '__main__':
    crawl_images(input('원하는 검색어: '))
```
